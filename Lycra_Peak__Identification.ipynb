```python
import matplotlib.pyplot as plt
import numpy as np
```

### Simple Example

```python
filename = r"Mitarth B20EE096_ BTP Data\Mitarth B20EE096_ BTP Data\Lycra\It-5s, LP-25_Lycra1.csv"
```

```python
with open(filename) as f:
    lines = f.readlines()
    x1 = []
    y1 = []
    for line in lines[99:]:
        currentline = line.split(",")
        if float(currentline[3])>0 :
            x1.append(int(float(currentline[3])))
            y1.append(int(float(currentline[7])))
            print(f"({int(float(currentline[3]))},{int(float(currentline[7]))})")
```

```python
type(x1[0])
```

```python
x1 = np.array(x1)
y1 = np.array(y1)
```

```python
plt.plot(x1,y1)
plt.show()
```

```python
for i in range(len(x1)):
    if x1[i]>500:
        break
for j in range(len(x1)):
    if x1[j]>2000:
        break
x_cut = x1[i:j]
y_cut = y1[i:j]
#x_cut
```

```python
len(x_cut)
```

```python
plt.plot(x_cut,y_cut)
plt.show()
```

```python
import scipy
```

```python
peaks = scipy.signal.find_peaks(y_cut, distance=50, threshold=10)
peaks[0]
```

```python
x_peaks = []
for i in peaks[0]:
    if i<1500:
        #print(x_cut[i])
        x_peaks.append(x_cut[i])
x_peaks
```

```python
y_peaks = []
for i in peaks[0]:
    y_peaks.append(y_cut[i])
y_peaks
```

```python
plt.plot(x_cut, y_cut)
plt.scatter(x_peaks, y_peaks)
plt.show()
```

```python
y_peaks_copy = y_peaks.copy()
max4_index = []
for i in range(4):
    max_value = max(y_peaks_copy)
    index = y_peaks_copy.index(max_value)
    max4_index.append(index)
    y_peaks_copy[index] = 100
    print(y_peaks[index])
max4_index
```

```python
x_max4=[]
y_max4=[]

for index in max4_index:
    x_max4.append(x_peaks[index])
    y_max4.append(y_peaks[index])
    print(index)
    print(y_peaks[index])
```

```python
plt.plot(x_cut, y_cut)
plt.scatter(x_max4,y_max4 , color='r')
plt.show()
```
